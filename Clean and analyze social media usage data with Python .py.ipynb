# Task-1 : Import required libraries


```python
# Task 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

# Task-2 : Generate random data

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
data = {
    'Date': pd.date_range('2021-01-01', periods=500),
    'Category': [random.choice(categories) for _ in range(500)],
    'Likes': np.random.randint(0, 10000, size=500)
}
```

# Task-3 : Load data into a Pandas DataFrame and explore

```python

df = pd.DataFrame(data)
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df['Category'].value_counts())
```

# Task-4 :  Clean the data

```python
# Task 4: Clean the data
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df['Likes'] = df['Likes'].astype(int)
```

# Task-5 : Visualize and analyze the data

```python
# Task 5: Visualize and analyze the data
sns.histplot(df['Likes'])
plt.title("Histogram of likes")
plt.show()
```

```python
sns.boxplot(x='Category', y='Likes', data=df)
plt.title("Box plot of Likes by category")
plt.show()
```

```python
print(df['Likes'].mean())
print(df.groupby('Category')['Likes'].mean())
```
